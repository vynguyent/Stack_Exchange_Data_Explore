```python
from pymongo import MongoClient
from bson.code import Code
client = MongoClient('0.tcp.ngrok.io:14162')
client.admin.authenticate("yida", "yida")

database_list = client.database_names()
database_list.pop(database_list.index("admin"))
database_list.pop(database_list.index("local"))
database_list
len(database_list)
```

```python
tag_list = {}
for community in database_list:
    print "'"+community+"'"+","
    posts = client[community]['Posts']
    for post in posts.find({"@Tags": {"$exists": True}}):
        tags = post["@Tags"]
        tags = tags.replace("><", ",").replace(">", "").replace("<", "")
        for tag in tags.split(","):
            tag_list[tag] = tag_list[tag] + 1 if tag in tag_list.keys() else 1
```
